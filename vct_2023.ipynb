```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import funcoes as fc
```

Importando o dataset de agents pick do vct 2023



```python
agents_vct2023= pd.read_csv('dataset/vct_2023/agents/agents_pick_rates.csv')
agents_vct2023
#A ideia é pegar apenas o "All Maps" e o "All Match types" de cada região
```

```python
math_type = "All Match Types"
map = "All Maps"
stage = "All Stages"
```

Dados do Champions 

```python
#Champions 2023
agents_vct2023_champion = agents_vct2023[(agents_vct2023['Tournament'] == "Valorant Champions 2023") & (agents_vct2023['Stage']==stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_champion_2023 = agents_vct2023_champion[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_champion_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_champion_2023
```

Dados Pacífico

```python
#Asia-Pacific
agents_vct2023_pacific = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: Pacific Last Chance Qualifier") & (agents_vct2023['Stage'] == stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_pacific_2023 = agents_vct2023_pacific[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_pacific_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_pacific_2023
```

Dados America

```python
#North-America
agents_vct2023_america = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: Americas Last Chance Qualifier") & (agents_vct2023['Stage']==stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_america_2023 = agents_vct2023_america[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_america_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_america_2023
```

Dados EMEA

```python
#Emea
agents_vct2023_emea = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: EMEA Last Chance Qualifier") & (agents_vct2023['Stage']== stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_emea_2023 = agents_vct2023_emea[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_emea_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_emea_2023
```

```python
#Retirando o % do Pick Rate e transformando em float
for i in [agente_pick_rate_champion_2023, agente_pick_rate_pacific_2023, agente_pick_rate_america_2023, agente_pick_rate_emea_2023]:
    
    #AVISO: esse i.loc[:,] foi indicação do Python para evitar um SettingWithCopyWarning e foi correção automática (também não entendi muito bem, mas funcionou)
    i.loc[:,'Pick Rate'] = pd.to_numeric(i['Pick Rate'].str.replace('%', ''), errors='coerce')
    
    #Transformando o Pick Rate em decimal
    i.loc[:,'Pick Rate'] = i['Pick Rate']/4

    
```

```python
from funcoes import juntar_tabelas
lista = [agente_pick_rate_champion_2023, agente_pick_rate_pacific_2023, agente_pick_rate_america_2023, agente_pick_rate_emea_2023]

agents_vct2023 = juntar_tabelas(lista, 'Agent', 'Pick Rate')
agents_vct2023
```

```python
agents_pick_vct2023 = pd.DataFrame(agents_vct2023, columns =['Agent', 'Pick Rate'])
caminho = 'datasets-criados/agents_pick_vct2023.csv'
agents_pick_vct2023.to_csv(caminho, index=False)
```

```python
#Plotando o gráfico de colunas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,10))
sns.barplot(x='Agent', y='Pick Rate', data=agents_pick_vct2023, order=agents_pick_vct2023.sort_values('Pick Rate', ascending=False)['Agent'])
plt.show()
```

```python
maps_stats_geral = pd.read_csv('dataset/vct_2023/agents/maps_stats.csv')
```

```python
maps_vct2023_champions = maps_stats_geral[(maps_stats_geral['Tournament'] == "Valorant Champions 2023") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2023_champions =  maps_vct2023_champions[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2023_champions.reset_index(drop=True, inplace=True)

maps_pick_vct2023_champions
```

```python
maps_vct2023_pacific = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour 2023: Pacific Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2023_pacific = maps_vct2023_pacific[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2023_pacific = pd.concat([maps_pick_vct2023_pacific, pd.DataFrame([{'Map':"Bind", "Attacker Side Win Percentage":'49%', "Defender Side Win Percentage":'51%'}])], ignore_index=True)

maps_pick_vct2023_pacific.reset_index(drop=True, inplace=True)

maps_pick_vct2023_pacific
```

```python
maps_vct2023_america = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour 2023: Americas Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2023_america = maps_vct2023_america[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2023_america.reset_index(drop=True, inplace=True)

maps_pick_vct2023_america
```

```python
maps_vct2023_emea = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour 2023: EMEA Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2023_emea = maps_vct2023_emea[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2023_emea.reset_index(drop=True, inplace=True)

maps_pick_vct2023_emea
```

```python
for maps_vct2023 in [maps_pick_vct2023_champions, maps_pick_vct2023_america, maps_pick_vct2023_emea, maps_pick_vct2023_pacific]:
    
    #AVISO: esse i.loc[:,] foi indicação do Python para evitar um SettingWithCopyWarning e foi correção automática (também não entendi muito bem, mas funcionou)
    maps_vct2023.loc[:,'Attacker Side Win Percentage'] = pd.to_numeric(maps_vct2023['Attacker Side Win Percentage'].str.replace('%', ''), errors='coerce')
    maps_vct2023.loc[:,'Defender Side Win Percentage'] = pd.to_numeric(maps_vct2023['Defender Side Win Percentage'].str.replace('%', ''), errors='coerce')



    #Transformando o Pick Rate em decimal
    maps_vct2023.loc[:,'Attacker Side Win Percentage'] = maps_vct2023['Attacker Side Win Percentage']/4
    maps_vct2023.loc[:,'Defender Side Win Percentage'] = maps_vct2023['Defender Side Win Percentage']/4
    
```

```python
from funcoes import juntar_tabelas
list_maps = [maps_pick_vct2023_champions, maps_pick_vct2023_emea, maps_pick_vct2023_america, maps_pick_vct2023_pacific]

ataque_2023 = juntar_tabelas(list_maps, 'Map', 'Attacker Side Win Percentage')
print(ataque_2023)
defesa_2023 = juntar_tabelas(list_maps, 'Map', 'Defender Side Win Percentage')
print(defesa_2023)

maps_vct2023 = pd.merge(ataque_2023, defesa_2023, on='Map')
#maps_vct2023
```

```python
caminho = 'datasets-criados/maps_vct2023.csv'
maps_vct2023.to_csv(caminho, index=False)
print(maps_vct2023)
```
