Importando de bibliotecas


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importando o dataset de agents pick do vct 2023



```python
agents_vct2023= pd.read_csv('dataset/vct_2023/agents/agents_pick_rates.csv')
agents_vct2023
#A ideia é pegar apenas o "All Maps" e o "All Match types" de cada região
```

```python
math_type = "All Match Types"
map = "All Maps"
stage = "All Stages"
```

Dados do Champions 

```python
#Champions 2023
agents_vct2023_champion = agents_vct2023[(agents_vct2023['Tournament'] == "Valorant Champions 2023") & (agents_vct2023['Stage']==stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_champion_2023 = agents_vct2023_champion[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_champion_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_champion_2023
```

Dados Pacífico

```python
#Asia-Pacific
agents_vct2023_pacific = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: Pacific Last Chance Qualifier") & (agents_vct2023['Stage'] == stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_pacific_2023 = agents_vct2023_pacific[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_pacific_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_pacific_2023
```

Dados America

```python
#North-America
agents_vct2023_america = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: Americas Last Chance Qualifier") & (agents_vct2023['Stage']==stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_america_2023 = agents_vct2023_america[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_america_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_america_2023
```

Dados EMEA

```python
#Emea
agents_vct2023_emea = agents_vct2023[(agents_vct2023['Tournament'] == "Champions Tour 2023: EMEA Last Chance Qualifier") & (agents_vct2023['Stage']== stage) & (agents_vct2023['Match Type'] == math_type) & (agents_vct2023['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_emea_2023 = agents_vct2023_emea[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_emea_2023.reset_index(drop=True, inplace=True)
agente_pick_rate_emea_2023
```

```python
#Retirando o % do Pick Rate e transformando em float
for i in [agente_pick_rate_champion_2023, agente_pick_rate_pacific_2023, agente_pick_rate_america_2023, agente_pick_rate_emea_2023]:
    
    #AVISO: esse i.loc[:,] foi indicação do Python para evitar um SettingWithCopyWarning e foi correção automática (também não entendi muito bem, mas funcionou)
    i.loc[:,'Pick Rate'] = pd.to_numeric(i['Pick Rate'].str.replace('%', ''), errors='coerce')
    
    #Transformando o Pick Rate em decimal
    i.loc[:,'Pick Rate'] = i['Pick Rate']/100

    
```

```python
#Pega a lista de agentes por um dos dataframes já criados
agentes = list(agente_pick_rate_champion_2023['Agent'])

#Criando um contador para cada agente
contadores = [0]*23

#Percorendo cada dataframe
for i in [agente_pick_rate_champion_2023, agente_pick_rate_pacific_2023, agente_pick_rate_america_2023, agente_pick_rate_emea_2023]:
    
    #Zipando os agentes com os pick rates
    for j in zip(i['Agent'], i['Pick Rate']):

        #Adicionando o pick rate ao contador do agente
        contadores[agentes.index(j[0])] += j[1]
            
print(contadores)


```

```python
#Criando um dataframe com os agentes e os pick rates final de todas as regiões
agents_pick_vct2023 = pd.DataFrame(list(zip(agentes, contadores)), columns =['Agent', 'Pick Rate'])
caminho = 'datasets-criados/agents_pick_vct2023.csv'
agents_pick_vct2023.to_csv(caminho, index=False)
print(agents_pick_vct2023)
```

```python
#Plotando o gráfico de colunas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,10))
sns.barplot(x='Agent', y='Pick Rate', data=agents_pick_vct2023, order=agents_pick_vct2023.sort_values('Pick Rate', ascending=False)['Agent'])
plt.show()
```
