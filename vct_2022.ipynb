Importando de bibliotecas


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importando o dataset de agents pick do vct 2022



```python
agents_vct2022= pd.read_csv('dataset/vct_2022/agents/agents_pick_rates.csv')
agents_vct2022
#A ideia é pegar apenas o "All Maps" e o "All Match types" de cada região
```

```python
math_type = "All Match Types"
map = "All Maps"
stage = "All Stages"
```

Dados do Champions 

```python
#Champions 2022
agents_vct2022_champion = agents_vct2022[(agents_vct2022['Tournament'] == "Valorant Champions 2022") & (agents_vct2022['Stage']==stage) & (agents_vct2022['Match Type'] == math_type) & (agents_vct2022['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_champion_2022 = agents_vct2022_champion[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_champion_2022.reset_index(drop=True, inplace=True)
agente_pick_rate_champion_2022
```

Dados Pacífico

```python
#Asia-Pacific
agents_vct2022_pacific = agents_vct2022[(agents_vct2022['Tournament'] == "Champions Tour Asia-Pacific: Last Chance Qualifier") & (agents_vct2022['Stage'] == stage) & (agents_vct2022['Match Type'] == math_type) & (agents_vct2022['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_pacific_2022 = agents_vct2022_pacific[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_pacific_2022.reset_index(drop=True, inplace=True)
agente_pick_rate_pacific_2022
```

Dados North America

```python
#North-America
agents_vct2022_north_america = agents_vct2022[(agents_vct2022['Tournament'] == "Champions Tour North America: Last Chance Qualifier") & (agents_vct2022['Stage']==stage) & (agents_vct2022['Match Type'] == math_type) & (agents_vct2022['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_north_america_2022 = agents_vct2022_north_america[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_north_america_2022.reset_index(drop=True, inplace=True)
agente_pick_rate_north_america_2022
```

Dados South America

```python
#South-America
agents_vct2022_south_america = agents_vct2022[(agents_vct2022['Tournament'] == "Champions Tour South America: Last Chance Qualifier") & (agents_vct2022['Stage']== stage) & (agents_vct2022['Match Type'] == math_type) & (agents_vct2022['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_south_america_2022 = agents_vct2022_south_america[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_south_america_2022.reset_index(drop=True, inplace=True)
agente_pick_rate_south_america_2022
```

Dados EMEA

```python
#Emea
agents_vct2022_emea = agents_vct2022[(agents_vct2022['Tournament'] == "Champions Tour EMEA: Last Chance Qualifier") & (agents_vct2022['Stage']== stage) & (agents_vct2022['Match Type'] == math_type) & (agents_vct2022['Map'] == map)]

#Pegando as colunas que interessam Agent e Pick Rate
agente_pick_rate_emea_2022 = agents_vct2022_emea[['Agent', 'Pick Rate']]

#Resetando o id
agente_pick_rate_emea_2022.reset_index(drop=True, inplace=True)
agente_pick_rate_emea_2022
```

```python
#Retirando o % do Pick Rate e transformando em float
for i in [agente_pick_rate_champion_2022, agente_pick_rate_pacific_2022, agente_pick_rate_north_america_2022, agente_pick_rate_south_america_2022, agente_pick_rate_emea_2022]:
    
    #AVISO: esse i.loc[:,] foi indicação do Python para evitar um SettingWithCopyWarning e foi correção automática (também não entendi muito bem, mas funcionou)
    i.loc[:,'Pick Rate'] = pd.to_numeric(i['Pick Rate'].str.replace('%', ''), errors='coerce')
    
    #Transformando o Pick Rate em decimal
    i.loc[:,'Pick Rate'] = i['Pick Rate']/5

    
```

```python
from funcoes import juntar_tabelas
lista = [agente_pick_rate_champion_2022, agente_pick_rate_pacific_2022, agente_pick_rate_north_america_2022, agente_pick_rate_south_america_2022, agente_pick_rate_emea_2022]

agents_vct2022 = juntar_tabelas(lista, 'Agent', 'Pick Rate')
agents_vct2022
```

```python
#Criando um dataframe com os agentes e os pick rates final de todas as regiões
agents_pick_vct2022 = pd.DataFrame(agents_vct2022, columns =['Agent', 'Pick Rate'])
caminho = 'datasets-criados/agents_pick_vct2022.csv'
agents_pick_vct2022.to_csv(caminho, index=False)
```

```python
#Plotando o gráfico de colunas
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,10))
sns.barplot(x='Agent', y='Pick Rate', data=agents_pick_vct2022, order=agents_pick_vct2022.sort_values('Pick Rate', ascending=False)['Agent'])
plt.show()
```

```python
maps_stats_geral = pd.read_csv('dataset/vct_2022/agents/maps_stats.csv')
```

```python
maps_vct2022_champions = maps_stats_geral[(maps_stats_geral['Tournament'] == "Valorant Champions 2022") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2022_champions = maps_vct2022_champions[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2022_champions.reset_index(drop=True, inplace=True)

maps_pick_vct2022_champions
```

```python
maps_vct2022_pacific = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour Asia-Pacific: Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2022_pacific = maps_vct2022_pacific[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2022_pacific.reset_index(drop=True, inplace=True)

maps_pick_vct2022_pacific
```

```python
maps_vct2022_north = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour North America: Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2022_north = maps_vct2022_north[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2022_north.reset_index(drop=True, inplace=True)

maps_pick_vct2022_north
```

```python
maps_vct2022_south = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour South America: Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2022_south = maps_vct2022_south[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2022_south.reset_index(drop=True, inplace=True)

maps_pick_vct2022_south
```

```python
maps_vct2022_emea = maps_stats_geral[(maps_stats_geral['Tournament'] == "Champions Tour EMEA: Last Chance Qualifier") & (maps_stats_geral['Stage']== stage) & (maps_stats_geral['Match Type'] == math_type)][1:]

maps_pick_vct2022_emea = maps_vct2022_south[['Map', 'Attacker Side Win Percentage', 'Defender Side Win Percentage']]

maps_pick_vct2022_emea.reset_index(drop=True, inplace=True)

maps_pick_vct2022_emea
```

```python
for maps_vct2022 in [maps_pick_vct2022_champions, maps_pick_vct2022_north, maps_pick_vct2022_emea, maps_pick_vct2022_pacific, maps_pick_vct2022_south]:
    
    #AVISO: esse i.loc[:,] foi indicação do Python para evitar um SettingWithCopyWarning e foi correção automática (também não entendi muito bem, mas funcionou)
    maps_vct2022.loc[:,'Attacker Side Win Percentage'] = pd.to_numeric(maps_vct2022['Attacker Side Win Percentage'].str.replace('%', ''), errors='coerce')
    maps_vct2022.loc[:,'Defender Side Win Percentage'] = pd.to_numeric(maps_vct2022['Defender Side Win Percentage'].str.replace('%', ''), errors='coerce')



    #Transformando o Pick Rate em decimal
    maps_vct2022.loc[:,'Attacker Side Win Percentage'] = maps_vct2022['Attacker Side Win Percentage']/5
    maps_vct2022.loc[:,'Defender Side Win Percentage'] = maps_vct2022['Defender Side Win Percentage']/5
    
```

```python
from funcoes import juntar_tabelas
list_maps = [maps_pick_vct2022_champions, maps_pick_vct2022_emea, maps_pick_vct2022_north, maps_pick_vct2022_pacific, maps_pick_vct2022_south]

ataque_2022 = juntar_tabelas(list_maps, 'Map', 'Attacker Side Win Percentage')
defesa_2022 = juntar_tabelas(list_maps, 'Map', 'Defender Side Win Percentage')

maps_vct2022 = pd.merge(ataque_2022, defesa_2022, on='Map')
maps_vct2022
```
